# 생선 분류문제 간단하게 해결하기
 - 목표: 도미와, 빙어 데이터를 토대로 도미와 빙어를 분류할 수 있는 기계학습 모델을 만들자
 - Input: 도미, 빙어의 길이와, 무게 as List
 - Output: 도미 or 빙어 ==>
 - 특성: 지도학습, 이진분류


```python
import matplotlib.pyplot as plt
import numpy as np
```

## 01. 데이터 가공

```python
#도미 데이터 불러오기
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0, 31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0, 35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0, 500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0, 700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
```

```python
bream_length[0], bream_weight[0]
```

### 01-1. 데이터 시각화

```python
plt.scatter(bream_length, bream_weight)
plt.xlabel("length")
plt.ylabel("weight")
plt.show()
```

```python
#빙어 데이터
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]
```

```python
plt.scatter(bream_length, bream_weight, c=['red'])
plt.scatter(smelt_length, smelt_weight, c=['blue'])
plt.xlabel("length")
plt.ylabel("weight")
plt.show()
```

```python
# X = 특징, Y = label
# X = 길이, 무게, Y = 도미 or 빙어

length = bream_length + smelt_length
weight = bream_weight + smelt_weight
```

```python
# "도미"==0, "방어"==1

fish_target = len(bream_length)*[0] + len(smelt_length)*[1]
```

```python
fish_target[38]
```

```python
length[38]
```

```python
weight[38]
```

```python
fish_data = [[l,w] for l,w in zip(length, weight)]
```

```python
fish_data
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
kn = KNeighborsClassifier()
```

```python
kn.fit(fish_data, fish_target)
```

```python
kn.score(fish_data, fish_target)
```

```python
plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.scatter(30, 600)
```

```python
kn.predict([(39.5, 925.0)])
```

```python
kn.predict(fish_data)
```

```python
fish_target
```

```python
kn = KNeighborsClassifier(n_neighbors=3)
```

```python
kn.fit(fish_data, fish_target)
```

```python
kn.score(fish_data, fish_target)
```

```python
train_input = fish_data[:35]
train_target = fish_target[:35]

test_input = fish_data[35:]
test_target = fish_target[35:]
```

```python
kn = kn.fit(train_input, train_target)
kn.score(test_input, test_target)
```
